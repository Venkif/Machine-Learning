# Data cleaning : 
# Removing age > 70 

```python
cr_age_removed = credit_risk[credit_risk['person_age']<=70]
cr_age_removed.reset_index(drop=True,inplace=True)
```

```python
cr_age_removed.head()
```

```python
cr_age_removed.shape
```

```python
cr_age_removed.head()
```

```python
cr_age_removed.pivot_table(index='person_emp_length',columns='loan_status',values='person_income',aggfunc='count').reset_index().sort_values(by ='person_emp_length',ascending=False)
```

# Removing work experience > 47 years

```python
person_emp_removed = cr_age_removed[cr_age_removed['person_emp_length'] <= 47]
person_emp_removed.reset_index(drop=True,inplace=True)
```

```python
person_emp_removed.shape
```

```python
person_emp_removed.describe()
```

# Checking null values

```python
person_emp_removed.isnull().sum()
```

```python
person_emp_removed.describe()
```

```python
cr_data = person_emp_removed.copy()
```

```python
cr_data.head()
```

# Filling the missing value with Median value since median and mean are nearly same.

```python
cr_data.fillna({'loan_int_rate' : cr_data['loan_int_rate'].median()},inplace=True)
```

```python
cr_data.isnull().sum()
```

# Droping loan grade

```python
cr_data_copy = cr_data.drop('loan_grade',axis=1)
```

```python
cr_data_copy.shape
```

# Handling categorical values: ONE HOT ENCODING

```python
cr_data_copy.groupby('loan_intent').count()['person_income']
```

```python
cr_data_copy.groupby('person_home_ownership').count()['person_income']
```

```python
cr_data_copy.head()
```

```python
cr_data_cat_treated = cr_data_copy.copy()
```

```python
cr_data_cat_treated.head()
```

```python
person_home_ownership = pd.get_dummies(cr_data_cat_treated['person_home_ownership'],drop_first=True).astype(int)
```

```python
person_home_ownership.head()
```

```python
loan_intent = pd.get_dummies(cr_data_cat_treated['loan_intent'],drop_first=True).astype(int)
```

```python
loan_intent.head()
```

```python
cr_data_cat_treated['cb_person_default_on_file_binary'] = np.where(cr_data_cat_treated['cb_person_default_on_file']=='Y',1,0)
```

```python
cr_data_cat_treated.head()
```

# Standardization of data : Scaling

### Removing the categerical columns 

```python
data_to_scale = cr_data_cat_treated.drop(['person_home_ownership','loan_intent','loan_status','cb_person_default_on_file','cb_person_default_on_file_binary'],axis=1)
```

```python
data_to_scale.head()
```

### (x-mean of x)/(std of x) formula for scaling

```python
scaler = StandardScaler()
```

```python
scaled_data = scaler.fit_transform(data_to_scale)
```

```python
data_to_scale.columns
```

### Array to Data frame

```python
scaled_df = pd.DataFrame(scaled_data,columns=['person_age', 'person_income', 'person_emp_length', 'loan_amnt',
       'loan_int_rate', 'loan_percent_income', 'cb_person_cred_hist_length'])
```

```python
scaled_df.head()
```

### Validating Scaled data

```python
round(np.mean(scaled_df.person_income),2)
```

```python
round(np.std(scaled_df.person_income),2)
```

### Concating all the lists together

```python
scaled_data_combined = pd.concat([scaled_df,person_home_ownership,loan_intent],axis=1)
```

```python
scaled_data_combined.head()
scaled_data_combined.shape
```

```python
scaled_data_combined['cb_person_default_on_file'] = cr_data_cat_treated['cb_person_default_on_file_binary']
```

```python
scaled_data_combined['loan_status'] = cr_data_cat_treated['loan_status']
```

```python
scaled_data_combined.head()
```
